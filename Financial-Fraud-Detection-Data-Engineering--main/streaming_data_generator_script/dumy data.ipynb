```python
import json
import random
import time
from datetime import datetime
from azure.storage.blob import BlobServiceClient

# Azure Storage Details
connection_string = "this was also mine bro, again use your own🙂"
container_name = "fraud-data-lake"

# Blob Client
blob_service_client = BlobServiceClient.from_connection_string(connection_string)

# Random Data
locations = ["Mumbai", "Delhi", "Bangalore", "Hyderabad"]
devices = ["mobile", "web", "ATM"]

def generate_transaction():
    return {
        "transaction_id": f"TXN{random.randint(100000,999999)}",
        "user_id": f"U{random.randint(1,500)}",
        "amount": random.randint(100, 50000),
        "location": random.choice(locations),
        "device": random.choice(devices),
        "timestamp": str(datetime.now())
    }

while True:
    
    transaction = generate_transaction()

    file_name = f"transaction_{int(time.time())}.json"

    blob_path = f"bronze/streaming/{file_name}"

    blob_client = blob_service_client.get_blob_client(
        container=container_name,
        blob=blob_path
    )

    blob_client.upload_blob(
        json.dumps(transaction),
        overwrite=True
    )

    print(f"Uploaded: {file_name}")
    print(transaction)

    time.sleep(5)
```
